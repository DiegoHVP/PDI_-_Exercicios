```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("linha_diferenca.csv")

# Pega colunas
indices = data["Pixel"]
valores_png = data["Valor_PNG"]
valores_yml = data["Valor_YML"]
diferencas = data["Diferenca"]


plt.figure(figsize=(10, 6))

plt.plot(indices, valores_png, label="Valores PNG", linestyle='-.', color="green")
plt.plot(indices, valores_yml, label="Valores YML", linestyle='dotted', color="blue")
plt.plot(indices, diferencas, label="Diferenca", color="red")

plt.title("Comparao entre PNG e YML")
plt.xlabel("Pixeis da linha")
plt.ylabel("Valores")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
